```python
# BOX PLOT

sns.boxplot(x='day', y="total_bill", data = tips_Df, hue="sex" )
plt.legend(loc="upper right")
```

```python
# VIOLIN PLOT
# this is the combination of box plot and kde plot 
sns.violinplot(x="day", y ="total_bill", data= tips_Df, hue="sex")
```

```python
sns.violinplot(x="day", y ="total_bill", data= tips_Df, hue="sex", split=True)
```

```python
# STRIP PLOT
# basically strip plot is going to draw a scatter plot representing all the data points 
# where one variable is categorical. 
# and we often use this to show different observations with boxplot 
# which is gonna represent the average distributions

plt.figure(figsize=(10,8))
sns.stripplot(x="day",
              y = "total_bill",
              data = tips_Df,
              jitter = True,
              hue="sex",
              dodge=True
             )

```

```python
# Swarm plot 
# combines the capabilities of multiple other plots

sns.violinplot(x = "day",
               y = "total_bill",
               data = tips_Df,
              )
sns.swarmplot(
               x = "day",
               y = "total_bill",
               data = tips_Df,
               color = "white"
)

```

```python
plt.figure(figsize=(8,8))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x = "day",
              y = "total_bill",
              data = tips_Df,
              hue = "sex",
              palette = "magma" # color names from matplotlib library in colormaps.html
             )
plt.legend(loc = 8)

# best - 0 , upper right -1 , upper left - 2, lower left - 3, lower right - 4, 
# right - 5, center left -  6, center right - 7, lower center - 8, 
# upper center - 9, center - 10
```

```python
## pair grid
# Subplot grid for plotting pairwise relationships in a dataset.
# This object maps each variable in a dataset onto a column and row 
# in a grid of multiple axes. Different axes-level plotting functions 
# can be used to draw bivariate plots in the upper and lower triangles,
# and the the marginal distribution of each variable can be shown on 
# the diagonal.

# Several different common plots can be generated in a single line 
# using pairplot(). Use PairGrid when you need more flexibility.


iris_g = sns.PairGrid(iris, hue = 'species')
# iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)  ## for diagonal 
iris_g.map_offdiag(plt.scatter) # for non diagonal 
iris_g.map_upper(plt.scatter) # for non diagonal upper part
iris_g.map_lower(sns.kdeplot) # or non diagonal lower part


```

```python
iris_g = sns.PairGrid(iris, hue="species",
                     x_vars = [
                        'sepal_length', 'sepal_width'
                     ], 
                     y_vars = [
                         'petal_length', 'petal_width'
                     ])
iris_g.map(plt.scatter)
iris_g.add_legend()
```
