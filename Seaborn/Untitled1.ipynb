```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('/home/vy/Data_Visualization/Matplotlib/data_for_time_series_plot.csv')
data.head()
```

```python
# line graph 

ax = sns.lineplot(x = 'Date',
                  y = 'Open',
                  data = data,
                  palette = 'magma'
                 )
# sns.set(rc={'figure.figsize':(25,20)})
# plt.figure(figsize=(10,10))
```

```python
data[['Volume', 'Date']].groupby('Date').agg({'Volume':'mean'})
```

```python
 ax = sns.barplot(x = 'Date',
                  y = 'Volume',
                  data = data,
#                   hue = 'Date',
#                   palette = "Blues",
#                   color = 'green'
                  )
```
