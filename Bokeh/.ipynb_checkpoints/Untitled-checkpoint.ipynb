```python
import bokeh
from bokeh.io import output_notebook, show
from bokeh.plotting import figure

```

```python
output_notebook()
```

```python
p = figure(plot_width = 400, plot_height = 400)

x = [1,2,3]
y = [3,4,5]

p.circle(x, y)

p.scatter(x, y, size = 20, )


show(p)
```

```python
p.line(x,y,line_width = 4, color = 'red', alpha = 0.5)
show(p)
```

```python
p.diamond(x,y, size = 50)
show(p)
```

```python
p = figure(plot_width = 400, plot_height = 400)

x = [1,2,3]
y = [3,4,5]

p.circle(x, y)
p.asterisk(x,y, size = 50)
show(p)
```

```python
p = figure(plot_width = 400, plot_height = 400)

x = [1,2,3]
y = [3,4,5]

p.circle(x, y)
p.triangle(x,y, size = 50)
show(p)
```

```python
df = pd.read_csv('/home/vy/Downloads/Caste.csv', parse_dates = ['year'])
df
```

```python
from bokeh.models import ColumnDataSource
```

```python
source = ColumnDataSource(df)
```

```python
source
```

```python
p = figure(plot_width = 400, plot_height = 400)
p.line(x = 'year', y = 'convicts', source = source)
show(p)
```

```python
p = figure(x_axis_type = 'datetime', plot_width = 400, plot_height = 400)
p.line(x = df['year'], y = df['convicts'],)
show(p)
```

```python
x_square = [20,5,15,4]
y_square = [4,10,3,17]
x_circle = [16 , 8, 24, 4]
y_circle = [5,10,15,20]
```

```python
p = figure(title = "Square and circle", plot_width = 400, plot_height = 400)

p.circle(x = x_circle, y = y_circle, color = 'green', size = 10)
p.square(x = x_square, y = x_square, color = "red", size = 10)
p.line(x = [20,30,10,14], y = [10,5,4,8], color = "navy")
show(p)
```

```python
x = ['positive', 'negative']
p = figure(plot_width = 600, plot_height = 600, x_range = x)
y = [60, 30]
color = ["green", 'red']
legend = ['positive', 'negative']
source = ColumnDataSource(dict(categories = x, top = y, color = color, legend = legend))
p.vbar(top = 'top',width = 0.4, x = 'categories',legend = 'legend', alpha = .5, color = 'color', source = source)
show(p)
```

```python
from bokeh.models import HoverTool
x = ['positive', 'negative']
p = figure(plot_width = 600, plot_height = 600, x_range = x, tools = ['hover', 'reset'])
y = [60, 30]
color = ["green", 'red']
legend = ['positive', 'negative']
hover = p.select(dict(type = HoverTool))
source = ColumnDataSource(dict(categories = x, top = y, color = color, legend = legend))
p.vbar(top = 'top',width = 0.4, x = 'categories',legend = 'legend', alpha = .5, color = 'color', source = source)
hover.tooltips = [('bar type', '@categories'), ('no of count', '@top')]
show(p)
```
